```python
df = df.drop(['story_score', 'acting_cast_score', 'music_score', 'rewatch_value_score','ep_watched', 'n_helpful'],axis = 1)
```

```python
df.head(10)
```

```python
m  = df['user_id'].shape[0]
uid = df['user_id'].tolist()
```

```python
count = 0 
for i in range(1, m):
    if uid[i] == uid[i-1]:
        df.loc[i, 'user_id'] = count
    else:
        count += 1
        df.loc[i, 'user_id'] = count
```

```python
df.loc[0, 'user_id'] = 0
```

```python
df2 = pd.read_csv('final_item.csv')
```

```python
df2.shape
```

```python
valid_titles = df2['drama_name'].unique()

df = df[df['title'].isin(valid_titles)]
```

```python
df.shape
```

```python
df = df.drop(['review_text'], axis = 1)
```

```python
df.head(30)
```

```python
ratings = df
genres = df2

# Merge (specify left and right column names)
merged = ratings.merge(genres, left_on='title', right_on='drama_name')

# Get genre columns (all except 'drama_name')
genre_cols = [col for col in genres.columns if col != ['drama_name','kdrama_id','pop','rank']]

# Calculate averages per user per genre
result = {}
for genre in genre_cols:
    avg = merged[merged[genre] == 1].groupby('user_id')['overall_score'].mean()
    result[genre] = avg

# Create dataframe
result_df = pd.DataFrame(result).reset_index()
```

```python
result_df.head()
```

```python
result_df = result_df.drop(['drama_name'], axis = 1)
```

```python
result_df = result_df.fillna(0)
```

```python
result_df.head(10)
```

```python
df_with_genres = df.merge(result_df, on='user_id', how='left')


```
